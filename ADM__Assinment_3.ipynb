```python
import pandas as pd
```

```python
df = pd.read_csv('/content/customer_purchases.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Annual_Income'], bins=10,edgecolor="black")
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.title('Distribution of Annual Income')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Purchase_Frequency'], df['Spending_Score'])
plt.xlabel('Purchase Frequency')
plt.ylabel('Spending Score')
plt.title('Purchase Frequency vs. Spending Score')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numerical_features = df[['Annual_Income', 'Spending_Score', 'Purchase_Frequency']]
correlation_matrix = numerical_features.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Spending_Score'])
plt.title('Box Plot of Spending Score')
plt.show()
```

```python
outliers = df[(df['Spending_Score'] < 10) | (df['Spending_Score'] > 90)]
print(outliers)
```
