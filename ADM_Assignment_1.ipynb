```python
import pandas as pd

# Assuming 'df' is your DataFrame loaded from 'Social_Media_Engagement.csv'

# Group numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Impute missing values
for col in numerical_cols:
    if df[col].isnull().any():  # Check if column has missing values
        if len(df[col].unique()) > 10:  # Use mean/median for continuous data
            # Check skewness to determine imputation method
            if df[col].skew() > 1: # Highly skewed
                df[col].fillna(df[col].median(), inplace=True) # Impute median for skewed data
            else:
                df[col].fillna(df[col].mean(), inplace=True) # Impute mean for relatively normal data
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)  # Mode for discrete data

for col in categorical_cols:
    if df[col].isnull().any():
        df[col].fillna(df[col].mode()[0], inplace=True)  # Mode for categorical data

print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming 'df' is your DataFrame and contains 'Age' and 'Salary' columns

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the selected columns
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])

print(df)
```

```python
def min_max_normalize(series):
  return (series - series.min()) / (series.max() - series.min())

df['Job_Satisfaction_Norm'] = min_max_normalize(df['Job_Satisfaction'])
df['Work_Hours_Per_Week_Norm'] = min_max_normalize(df['Work_Hours_Per_Week'])

print(df[['Job_Satisfaction', 'Job_Satisfaction_Norm',
'Work_Hours_Per_Week', 'Work_Hours_Per_Week_Norm']].head())
```
