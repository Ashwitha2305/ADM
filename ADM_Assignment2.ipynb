```python
import pandas as pd
df = pd.read_csv("employee_performance.csv")
df['Department_Encoded'] = pd.factorize(df['Department'])[0]
df['Annual_Salary'] = df['Monthly_Salary'] * 12
grouped_df = df.groupby('Department').agg(
    Avg_Performance_Score=('Performance_Score', 'mean'),
        Total_Annual_Salary=('Annual_Salary', 'sum'),
            Employee_Count=('Employee_ID', 'count')
            ).reset_index()
print("Transformed Dataset Sample:")
print(df.head())
print("\nGrouped and Aggregated Data:")
print(grouped_df.head())
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Monthly_Salary'], bins=10, edgecolor='black')
plt.xlabel('Monthly Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Monthly Salaries')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Years_of_Experience'], df['Performance_Score'])
plt.xlabel('Years of Experience')
plt.ylabel('Performance Score')
plt.title('Years of Experience vs. Performance Score')
plt.show()
```

```python
numerical_features = df[['Monthly_Salary', 'Years_of_Experience', 'Performance_Score']]
correlation_matrix = numerical_features.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Salary, Experience, and Performance')
plt.show()
```
